```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python

#driver = webdriver.Edge(executable_path="C:\\Users\\user\\Documents\\datascience\\msedgedriver.exe")
#driver = webdriver.Chrome("/usr/lib/chromium-browser/chromedriver")

baseurl = "https://brickseek.com"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}
```

```python
productlinks = []
t = {}
data = []
c = 0

k = requests.get('https://brickseek.com/office-depot-inventory-checker/').text
soup = BeautifulSoup(k, 'html.parser')
productlist = soup.find_all(
    "div", {"class": "item-list__item item-list__item--deal"})


for product in productlist:
    link = product.find("a", {"class": "item-list__link"}).get('href')
    productlinks.append(baseurl + link)
productlinks
```

```python
amazonlink = ''

for link in productlinks:
    f = requests.get(link, headers=headers).text
    hun = BeautifulSoup(f, 'html.parser')

    try:
        price = hun.find("span", {
                         "class": "price-formatted price-formatted--style-normal"}).text.replace('\n', "")
        print(price)
    except:
        price = None
        # print("iouu")

    try:
        about = hun.find(
            "h2", {"class": "item-overview__title"}).text.replace('\n', "")
    except:
        about = None
    try:
        amazonlink = hun.find(
            "a", {"class": "item-overview__online-listing"}).get('href')
        substring = "goto.walmart.com"

        if substring in amazonlink:
            amazonlink = 'not on amazon'
    except:
        about = None

    new = {"price": price, "about": about, "amazonlink": amazonlink}

    data.append(new)
    c = c+1
   # print("completed",c)

df = pd.DataFrame(data)

# print(df)
```
