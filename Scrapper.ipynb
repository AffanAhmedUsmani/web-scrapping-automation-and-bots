```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python

baseurl1 = "https://teespring.com/shop/funny-men-s-tshirts?page={}&tsmac=marketplace&tsmic=navigation"
baseurl2 = "https://www.redbubble.com/shop/*?iaCode=u-clothing&page={}&sortOrder=trending"

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'}
```

```python
TrendingApparelTexts = []

t={}
data=[]
c=0
for i in range(1,20):
    
    page = requests.get(baseurl1.format(i))
   
    soup=BeautifulSoup(page.content,'html.parser')
    productlist = soup.find_all("div",{"class":"product_card"})

    for product in productlist:
        link = product.find("div",{"class":"product_card__title"}).text
        TrendingApparelTexts.append(link)

TrendingApparelTexts
```

```python
TrendingApparelTexts2 =[]
for i in range(1,100):
    
    page = requests.get(baseurl2.format(i))

    soup=BeautifulSoup(page.content,'html.parser')
    productlist = soup.find_all("a",{"class":"styles__link--3QJ5N"})

    for product in productlist:
        link = product.find ("span",{"class":"styles__box--2Ufmy styles__text--23E5U styles__display6--3wsBG styles__nowrap--33UtL styles__display-block--3kWC4"})
        link = link.text
        TrendingApparelTexts2.append(link)

TrendingApparelTexts2
```

```python
s=['#','Mini Skirt','Skirt','Chiffon Top','T-Shirt','Classic','Classical','Essential','Short-Sleeve','shirt','Top','Premium','leggings','Classic','Pullover','Sweatshirt','Hoodie','Zipped']
finale =[]
for i in range(0,len(TrendingApparelTexts2)):
    processing = TrendingApparelTexts2[i]
   
    for j in range(0,len(s)):
        
        Delete = s[j]
        if Delete in processing:
            processing = processing.replace(Delete,'')
            
    
    finale.append(processing)
print(finale)       
```

```python
with open("k.txt", "w",encoding="utf-8") as f:
    
    
    for item in finale:
        f.write("%s\n" % item)
```
